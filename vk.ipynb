```python
# импорт библиотек
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Загрузка данных из файлов
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
songs_df = pd.read_csv('songs.csv')
members_df = pd.read_csv('members.csv')
song_extra_info_df = pd.read_csv('song_extra_info.csv')
```

```python
# Создание словаря для хранения датафреймов
dataframes = {
    'train_df': train_df,
    'test_df': test_df,
    'songs_df': songs_df,
    'members_df': members_df,
    'song_extra_info_df' : song_extra_info_df
}

# Вывод первых 5 строк каждого датафрейма
for df_name, df in dataframes.items():
    print(f"\n{df_name}:")
    display(df.head())
```

```python
# Получение общей информации о данных
for df_name, df in dataframes.items():
    print(f"\n{df_name}:")
    print(df.info())
```

```python
# Описательная статистика возраста
members_df['bd'].describe()
```

```python
# Анализ целевой переменной
target_counts = train_df['target'].value_counts()
target_counts
```

```python
# Визуализация данных
plt.figure(figsize=(10, 6))
sns.histplot(data=members_df, x='bd', bins=50, kde=True)
plt.title("Распределение по возрасту")
plt.xlabel("Age")
plt.ylabel("Count")
plt.xlim(right=150)
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='target')
plt.title("Распределение целевой переменной")
plt.xlabel("Target")
plt.ylabel("Count")
plt.show()
```
